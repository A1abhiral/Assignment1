```python
# Problem 1: Getting Started with Data Exploration
# 1. Data Exploration and Understanding:

import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/assignment 1/Copy of WHR-2024-5CS037.csv")

# • Dataset Overview:
# 1. Load the dataset and display the first 10 rows.
print("First 10 rows: ")
print(data.head(10))
print("\n")

# 2. Identify the number of rows and columns in the dataset.
rows, cols = data.shape
print(f"\nNumber of rows: {rows}")
print(f"Number of columns: {cols}")

# 3. List all the columns and their data types.
print("\nColumns and their data types:")
print(data.dtypes)

# • Basic Statistics:
# 1. Calculate the mean, median, and standard deviation for the Score column.
mean = data['score'].mean()
median = data['score'].median()
std_score = data['score'].std()
print(f"\nMean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {std_score}")
print("\n")

# 2. Identify the country with the highest and lowest happiness scores.

# country with the highest happiness score
highest_index = data['score'].idxmax()
highest_country = data.loc[highest_index, 'Country name']
highest_score = data.loc[highest_index, 'score']

# country with the lowest happiness score
lowest_index = data['score'].idxmin()
lowest_country = data.loc[lowest_index, 'Country name']
lowest_score = data.loc[lowest_index, 'score']

print(f"Country with the highest happiness score: {highest_country} with score {highest_score}")
print(f"Country with the lowest happiness score: {lowest_country} with score {lowest_score}")
print("\n")

# • Missing Values:
# 1. Check if there are any missing values in the dataset. If so, display the total count for each column.
missing_values = data.isnull().sum()
print("Missing values coubt for each column:")
print(missing_values)
print("\n")

# • Filtering and Sorting:
# 1. Filter the dataset to show only the countries with a Score greater than 7.5.
filtered = data[data['score'] > 7.5]

# 2. For the filtered dataset - Sort the dataset by GDP per Capita in descending order and display the top 10 rows.
sorted = filtered.sort_values(by='Log GDP per capita', ascending=False)
print("Top 10 sorted dataset by GDP per Capita:")
print(sorted.head(10))
print("\n")

#• Adding New Columns:
#1. Create a new column called Happiness Category that categorizes countries into three categories based on their Score.
import numpy as np

condition = [
    (data['score']<4),
    (data['score']>=4) & (data['score'] <= 6),
    (data['score']>6)
]

choice = ['low','medium','high']

data['Happiness Category'] = np.select(condition, choice, default='unknown')

print(data[['Country name', 'score', 'Happiness Category']].head(5))
```

```python
# Problem - 2 - Some Advance Data Exploration Task:
# Task - 1 - Setup Task - Preparing the South-Asia Dataset:
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/content/drive/MyDrive/assignment 1/Copy of WHR-2024-5CS037.csv")

# Step 1: Define the countries in South Asia
south_asian_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India",
                         "Maldives", "Nepal", "Pakistan", "Srilanka"]

#Step 2: filter out matching dataset from list.
south_asia_data = data[data['Country name'].isin(south_asian_countries)]

#Step 3: Save the filtered dataframe as separate CSV files for future use.
south_asia_data.to_csv('/content/drive/MyDrive/assignment 1/south_asia_WHR_data.csv', index=False)


#Task - 2 - Composite Score Ranking:
# 1. Using the SouthAsia DataFrame, create a new column called Composite Score

SA_data = pd.read_csv('/content/drive/MyDrive/assignment 1/south_asia_WHR_data.csv')

SA_data['Composite Score'] = (0.40*SA_data['Log GDP per capita'] +
                                      0.30*SA_data['Social support'] +
                                      0.30*SA_data['Healthy life expectancy'] )

# 2. Rank the South Asian countries based on the Composite Score in descending order.
SA_data = SA_data.sort_values(by='Composite Score', ascending=False)

# Visualize the top 5 countries using a horizontal bar chart showing the Composite Score.
top5_countries = SA_data.head(5)

plt.figure(figsize=(10, 6))
plt.barh(top5_countries['Country name'], top5_countries['Composite Score'], color='skyblue')
plt.xlabel('Composite Score')
plt.ylabel('Country')
plt.title('Top 5 South Asian countries by Composite Score')
plt.show()
print("\n")

# Step 4: Compare the rankings of Composite Score and Original Score
# Sort by the original score in descending order
SA_data_sorted_by_score = SA_data.sort_values(by='score', ascending=False)

# Plotting the rankings for Composite Score vs. Original Score
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Composite Score Plot
ax[0].barh(SA_data['Country name'], SA_data['Composite Score'], color='skyblue')
ax[0].set_xlabel('Composite Score')
ax[0].set_ylabel('Country')
ax[0].set_title('Ranking Based on Composite Score')

# Original Score Plot
ax[1].barh(SA_data_sorted_by_score['Country name'], SA_data_sorted_by_score['score'], color='lightcoral')
ax[1].set_xlabel('Original Score')
ax[1].set_ylabel('Country')
ax[1].set_title('Ranking Based on Original Score')

plt.tight_layout()
plt.show()
print("\n")


# Task - 3 - Outlier Detection:
# Define outliers using the 1.5 × IQR rule.
def outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# Identify outlier countries in South Asia based on their Score and GDP per Capita.
outlier_score = outliers(SA_data, 'score')
outlier_gdp = outliers(SA_data, 'Log GDP per capita')

outlier = pd.concat([outlier_score, outlier_gdp]).drop_duplicates()

# Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers in a different color.
plt.figure(figsize=(10,6))
plt.scatter(SA_data['Log GDP per capita'], SA_data['score'], alpha=0.5, edgecolor='black')
#highlight outliers
plt.scatter(outlier['Log GDP per capita'], outlier['score'], color='red', label='Outliers')
plt.xlabel('GDP per Capita')
plt.ylabel('Score')
plt.title('Outlier Detection in South Asia')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

Outlier countries may have high or low GDP per Capita / socres different from other regional countries. This outliers can give us wrong information by inflating regional average.

```python
#Problem 2
# Task - 4 - Exploring Trends Across Metrics:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

SA_data = pd.read_csv('/content/drive/MyDrive/assignment 1/south_asia_WHR_data.csv')

# 1. Choose two metrics (e.g., Freedom to Make Life Choices and Generosity) and calculate their correlation {pearson correlation} with the Score for South Asian countries.
correlation_freedom = SA_data['Freedom to make life choices'].corr(SA_data['score'], method='pearson')
correlation_generosity = SA_data['Generosity'].corr(SA_data['score'], method='pearson')

print(f"Correlation between Freedom to Make Life Choices and Score: {correlation_freedom}")
print(f"Correlation between Generosity and Score: {correlation_generosity}")

# 2. Create scatter plots with trendlines for these metrics against the Score.
# For Freedom to Make Life Choices
plt.figure(figsize=(10, 6))
sns.regplot(x='Freedom to make life choices', y='score', data=SA_data, scatter_kws={'alpha': 0.5}, line_kws={'color': 'red'})
plt.xlabel('Freedom to Make Life Choices')
plt.ylabel('Score')
plt.title('Trendline for Freedom to Make Life Choices vs. Score')
plt.show()
# For Generosity
plt.figure(figsize=(10, 6))
sns.regplot(x='Generosity', y='score', data=SA_data, scatter_kws={'alpha': 0.5}, line_kws={'color': 'red'})
plt.xlabel('Generosity')
plt.ylabel('Score')
plt.title('Trendline for Generosity vs. Score')
plt.show()
```

Since both the corellation is positive and strong (close to 1) it suggests that when x axis increases y axis also increases i.e. countries where people have more feedom they tend to have higher happiness score and generous countries tend to have higher happiness scores.

```python
# Problem 2
# Task - 5 - Gap Analysis:
import pandas as pd
import matplotlib.pyplot as plt

SA_data = pd.read_csv('/content/drive/MyDrive/assignment 1/south_asia_WHR_data.csv')

#1. Add a new column, GDP-Score Gap, which is the difference between GDP per Capita and the Score for each South Asian country.
SA_data['GDP-Score Gap'] = SA_data['Log GDP per capita'] - SA_data['score']

# 2. Rank the South Asian countries by this gap in both ascending and descending order.
SA_data_asc = SA_data.sort_values(by='GDP-Score Gap', ascending=True)
SA_data_desc = SA_data.sort_values(by='GDP-Score Gap', ascending=False)

# 3. Highlight the top 3 countries with the largest positive and negative gaps using a bar chart.
top3_asc = SA_data_asc.head(3)
top3_desc = SA_data_desc.head(3)

combined_gaps = pd.concat([top3_asc, top3_desc])
plt.figure(figsize=(10, 6))
plt.barh(combined_gaps['Country name'], combined_gaps['GDP-Score Gap'],
        color=['green' if x > 0 else 'red' for x in combined_gaps['GDP-Score Gap']])
plt.xlabel('GDP-Score Gap')
plt.ylabel('Country')
plt.title('Top 3 Countries with Largest Positive and Negative Gaps')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

These countries emphasizes on life satisfaction over material wealth because they have lower GDP but higher satisfaction.

```python

# Problem - 3 - Comparative Analysis:
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("/content/drive/MyDrive/assignment 1/Copy of WHR-2024-5CS037.csv")
SA_data = pd.read_csv('/content/drive/MyDrive/assignment 1/south_asia_WHR_data.csv')

# Task - 1 - Setup Task - Preparing the Middle Eastern Dataset:
middle_east_countries = [ "Bahrain", "Iran", "Iraq", "Israel", "Jordan",
"Kuwait", "Lebanon", "Oman", "Palestine", "Qatar", "Saudi Arabia", "Syria","United Arab Emirates", "Yemen"]

ME_data = data[data['Country name'].isin(middle_east_countries)]

ME_data.to_csv('/content/drive/MyDrive/assignment 1/middle_east_WHR_data.csv', index=False)


# 1. Descriptive Statistics:
# • Calculate the mean, Standard deviation of the score for both South Asia and Middle East.
SA_mean = SA_data['score'].mean()
SA_std = SA_data['score'].std()

ME_mean = ME_data['score'].mean()
ME_std = ME_data['score'].std()

print(f"South Asia - Mean happiness score: {SA_mean:.2f}, Standard Deviation: {SA_std:.2f}")
print(f"Middle East - Mean happiness score: {ME_mean:.2f}, Standard Deviation: {ME_std:.2f}")
print("\n")

# • Which region has higher happiness Scores on average?
if SA_mean > ME_mean:
    print("South Asia has a higher average happiness score.")
elif SA_mean < ME_mean:
    print("Middle East has a higher average happiness score.")
else:
    print("Both regions have the same average happiness score.")
print(" ")

# 2. Top and Bottom Performers:
# • Identify the top 3 and bottom 3 countries in each region based on the score.
SA_top3 = SA_data.sort_values(by='score', ascending=False).head(3)
SA_bottom3 = SA_data.sort_values(by='score', ascending=True).head(3)

ME_top3 = ME_data.sort_values(by='score', ascending=False).head(3)
ME_bottom3 = ME_data.sort_values(by='score', ascending=True).head(3)

print("Top 3 countries in South Asia:")
print(SA_top3[['Country name', 'score']])
print("\nBottom 3 countries in South Asia:")
print(SA_bottom3[['Country name', 'score']])

print("top 3 countries in Middle East:")
print(ME_top3[['Country name', 'score']])
print("\nBottom 3 countries in Middle East:")
print(ME_bottom3[['Country name', 'score']])

# • Plot bar charts comparing these charts.
SA_combined = pd.concat([SA_top3, SA_bottom3])
ME_combined = pd.concat([ME_top3, ME_bottom3])

# Plot for South Asia
plt.figure(figsize=(12, 6))
plt.bar(SA_combined['Country name'], SA_combined['score'], color='skyblue')
plt.title('Top and Bottom Performers in South Asia (Happiness Score)')
plt.ylabel('Happiness Score')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot for Middle East
plt.figure(figsize=(12, 6))
plt.bar(ME_combined['Country name'], ME_combined['score'], color='lightcoral')
plt.title('Top and Bottom Performers in Middle East (Happiness Score)')
plt.ylabel('Happiness Score')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. Metric Comparisons:
# • Compare key metrics like GDP per Capita, Social Support, and Healthy Life Expectancy between the regions using grouped bar charts.
SA_metrics_mean = SA_data[['Log GDP per capita', 'Social support', 'Healthy life expectancy']].mean()
ME_metrics_mean = ME_data[['Log GDP per capita', 'Social support', 'Healthy life expectancy']].mean()

# Combine into a DataFrame for comparison
metrics_comparison = pd.DataFrame({
    'South Asia': SA_metrics_mean,
    'Middle East': ME_metrics_mean
}).T

# Plot grouped bar chart
metrics_comparison.plot(kind='bar', figsize=(10, 6), color=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Comparison of Key Metrics Between South Asia and Middle East')
plt.ylabel('Average Value')
plt.xlabel('Region')
plt.xticks(rotation=0)
plt.legend(title='Metrics')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
print("\n")


# Descriptive statistics for disparity
disparity = abs(SA_mean - ME_mean)
print(f"Disparity in happiness scores: {disparity:.2f}")

# Task 4: Happiness Disparity
range_south_asia = SA_data['score'].max() - SA_data['score'].min()
cv_south_asia = SA_std / SA_mean

range_middle_east = ME_data['score'].max() - ME_data['score'].min()
cv_middle_east = ME_std / ME_mean

print(f"Range of South Asia: {range_south_asia:.2f}")
print(f"Coefficient of Variation of South Asia: {cv_south_asia:.2f}")
print(f"Range of Middle East: {range_middle_east:.2f}")
print(f"Coefficient of Variation of Middle East: {cv_middle_east:.2f}")

if range_south_asia > range_middle_east:
    print("South Asia has greater variability in happiness.")
else:
    print("Middle East has greater variability in happiness.")
print("\n")

# Task 5: Correlation of score with other metrics
sa_cor_freedom = SA_data['Freedom to make life choices'].corr(SA_data['score'])
sa_cor_generosity = SA_data['Generosity'].corr(SA_data['score'])

me_cor_freedom = ME_data['Freedom to make life choices'].corr(ME_data['score'])
me_cor_generosity = ME_data['Generosity'].corr(ME_data['score'])

print(f"Correlation in South Asia (Freedom to make life choices): {sa_cor_freedom:.2f}")
print(f"Correlation in South Asia (Generosity): {sa_cor_generosity:.2f}")
print(f"Correlation in Middle East (Freedom to make life choices): {me_cor_freedom:.2f}")
print(f"Correlation in Middle East (Generosity): {me_cor_generosity:.2f}")

# Scatter plots for South Asia
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.regplot(x='Freedom to make life choices', y='score', data=SA_data, color='green')
plt.title('Freedom vs Score in South Asia')
plt.xlabel('Freedom to make life choices')
plt.ylabel('Score')

plt.subplot(1, 2, 2)
sns.regplot(x='Generosity', y='score', data=SA_data, color='red')
plt.title('Generosity vs Score in South Asia')
plt.xlabel('Generosity')
plt.ylabel('Score')
plt.tight_layout()
plt.show()
print("\n")

# Task 6: Outlier Detection
def detect_outliers(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

sa_outliers = detect_outliers(SA_data, 'score')
me_outliers = detect_outliers(ME_data, 'score')

# Scatter plot for outliers in South Asia
sns.scatterplot(x="Log GDP per capita", y="score", data=SA_data, color="blue", label="Non-Outliers")
sns.scatterplot(x=sa_outliers["Log GDP per capita"], y=sa_outliers["score"], color="red", label="Outliers")
plt.title("Outliers Detection in South Asia")
plt.xlabel("Log GDP per capita")
plt.ylabel("Score")
plt.legend()
plt.show()
print("\n")

# Scatter plot for outliers in Middle East
sns.scatterplot(x="Log GDP per capita", y="score", data=ME_data, color="blue", label="Non-Outliers")
sns.scatterplot(x=me_outliers["Log GDP per capita"], y=me_outliers["score"], color="red", label="Outliers")
plt.title("Outliers Detection in Middle East")
plt.xlabel("Log GDP per capita")
plt.ylabel("Score")
plt.legend()
plt.show()
print("\n")

# Task 7: Boxplot Comparison
plt.figure(figsize=(10, 6))
sns.boxplot(data=[SA_data['score'], ME_data['score']], showmeans=True)
plt.xticks([0, 1], ["South Asia", "Middle East"])
plt.title("Comparison of Score Distribution")
plt.xlabel("Region")
plt.ylabel("Score")
plt.show()
```
